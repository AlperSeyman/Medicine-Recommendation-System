## Train Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X = data.drop("prognosis", axis=1)
y = data["prognosis"]
```

```python
X.head()
```

```python
y.head()
```

```python
le = LabelEncoder()
le.fit(y)
Y = le.transform(y)
```

```python
X_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
X_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## Training Top Models

```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
```

```python
models = {
    "SVC": SVC(kernel="linear"),
    "RandomForest": RandomForestClassifier(n_estimators=100,random_state=42),
    "GrandientBoostring": GradientBoostingClassifier(n_estimators=100, random_state=42),
    "KNeighbors": KNeighborsClassifier(n_neighbors=5),
    "MultinomialNB": MultinomialNB()
}
for model_name, model_algorithm in models.items():
    # train models
    model_algorithm.fit(X_train, y_train)
    
    # test models
    y_predictions = model_algorithm.predict(x_test)

    # accuracy
    accuracy = accuracy_score(y_test, y_predictions)

    print(f"{model_name} Accuracy: {accuracy:.2f}")
    
```

## Single Prediction

```python
svc = SVC(kernel="linear")
svc.fit(X_train, y_train)
y_pred = svc.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python


def filter(predicted_diseas):
    desrc = description[description["Disease"] == predicted_diseas]["Description"]
    desrc = " ".join(desrc) 

    pre = precautions[precautions["Disease"] == predicted_diseas][["Precaution_1","Precaution_2","Precaution_3","Precaution_4"]]
    pre = [col for col in pre.values]

    med = medications[medications["Disease"] == predicted_diseas]["Medication"]
    med = [col for col in med.values]

    diet = diets[diets["Disease"] == predicted_diseas]["Diet"]
    diet = [col for col in diet.values]

    work_out = workout[workout["disease"] == predicted_diseas]["workout"]

    return desrc, pre, med, diet, work_out


def get_predict_value(patient_symptoms):
    input_vector = np.zeros(len(symptoms_dict))

    for item in patient_symptoms:
        input_vector[symptoms_dict[item]] = 1
    return diseases_list[svc.predict([input_vector])[0]]
```

```python
user_systomps = input("Enter your symtops.......")
user_systomps = [s.strip() for s in user_systomps.split(',')]
user_systomps = [s.strip("[]' ") for s in user_systomps]
predicted_diseas = get_predict_value(user_systomps)
desrc, pre, medic, diet, work_out = filter(predicted_diseas)
```
